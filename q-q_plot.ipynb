```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate two samples from normal distribution
sample_size = 100
mu = 0
sigma = 1
sample1 = np.random.normal(mu, sigma, sample_size)
sample2 = np.random.normal(mu, sigma,  sample_size)
sample3 = np.random.exponential(0.2,  sample_size)

# Sort samples
sample1_sorted = np.sort(sample1)
sample2_sorted = np.sort(sample2)
sample3_sorted = np.sort(sample3)

# Calculate quantiles
p = np.arange(1, sample_size+1) / sample_size
q1 = np.quantile(sample1_sorted, p)
q2 = np.quantile(sample2_sorted, p)
q3 = np.quantile(sample3_sorted, p)

# Plot Q-Q plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
sns.histplot(q3, ax=ax1, bins=20, kde=True, stat='density')
ax2.scatter(q1, q3)
ax2.plot([q1.min(), q1.max()], [q1.min(), q1.max()], '--', color='gray')
ax2.set_xlabel('Sample 1 quantiles')
ax2.set_ylabel('Sample 2 quantiles')
plt.show()
```
