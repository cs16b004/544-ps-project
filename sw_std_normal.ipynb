```python
import numpy as np
from scipy.stats import shapiro
import seaborn as sns
import numpy as np
from scipy.stats import skewnorm, shapiro
```

```python
# generate 1000 samples from standard normal distribution
samples = np.random.normal(loc=0, scale=1, size=1000)
sns.histplot(samples, bins=20, kde=True, stat='density')

# perform Shapiro-Wilk test on the samples
stat, p = shapiro(samples)

# print the test statistic and p-value
print(f"Shapiro-Wilk test statistic: {stat:.4f}")
print(f"p-value: {p:.4f}")
```

```python


# generate 1000 samples from a left-skewed normal distribution
samples = skewnorm.rvs(a=-10, size=1000)
sns.histplot(samples, bins=20, kde=True, stat='density')

# perform Shapiro-Wilk test on the samples
stat, p = shapiro(samples)

# print the test statistic and p-value
print(f"Shapiro-Wilk test statistic: {stat:.4f}")
print(f"p-value: {p:.4f}")
```

```python
# generate 1000 samples from standard normal distribution
samples = np.random.exponential(0.2, size=1000)
sns.histplot(samples, bins=20, kde=True, stat='density')

# perform Shapiro-Wilk test on the samples
stat, p = shapiro(samples)

# print the test statistic and p-value
print(f"Shapiro-Wilk test statistic: {stat:.4f}")
print(f"p-value: {p:.4f}")
```
